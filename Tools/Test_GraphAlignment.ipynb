```python
import graph_alignment as ga

print(ga.__doc__)
```

```python
ref_seq = 'Cys(1)(2)--Cys--OH-DL-Val(2)--4OH-Leu--OH-Ile(1)'
query_seq = 'Cys(1)(2)--Cys--OH-DL-Val(2)--4OH-Leu--OH-Ile(1)'

ref_nodes, ref_edges = ga.sequence_to_node_edge(ref_seq)
ref_G = ga.create_graph(ref_nodes, ref_edges)
ga.plot_graph(ref_G, dpi=30)

query_nodes, query_edges = ga.sequence_to_node_edge(query_seq)
query_G = ga.create_graph(query_nodes, query_edges)
ga.plot_graph(query_G, dpi=30)

print('GED :', ga.graph_similarity(query_G, ref_G))
print('MCS :', ga.mcs_similarity(query_G, ref_G))
```
