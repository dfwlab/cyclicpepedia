#### Download data from ConoServer (http://www.conoserver.org/index.php)

```python
##import libraries
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
##***Extract cyclic peptides from ConoServer***##
##read html (all peptides)
peptide_html = BeautifulSoup(open("ConoServer_peptide.html"),'html.parser')
##extract tables
peptide_tables = peptide_html.find_all('table')
peptide_table = peptide_tables[1] ##id="datatable"
peptide_data = pd.read_html(peptide_table.prettify())
```

```python
##remove row/column only contains NAs
print(peptide_data[0].shape)
ConoData = peptide_data[0].dropna(axis=0,how='all')
ConoData = ConoData.dropna(axis=1,how='all')
print(ConoData.shape)
```

```python
ConoData.head()
```

```python
##save to excel
ConoData.to_excel('ConoServer_data.xlsx', index=False, header=False)
```

```python
##extract cyclic peptides
ConoData = pd.read_excel('ConoServer_data.xlsx')
print(ConoData.shape)
ConoData_cyclic = ConoData[~ConoData['Name'].str.contains('precursor', case=False)] #remover precursors
print(ConoData_cyclic.shape)

#super family: A/C/D/E/I1-3/J/K/L/M/O1-3/P/T/U
#cysteines frameworks: I;IV;V;VI/VII;IX;X;XI;XIV
ConoData_cyclic = ConoData_cyclic[(ConoData_cyclic['Gene superfamily'].str.contains('A |C |D |E |I1 |I2 |I3 |J |K |L |M |O1 |O2 |O3 |P |T |U ', case=True, na=False)) | 
                      (ConoData_cyclic['Cysteine framework'].isin(['I','IV','V','VI/VII','IX','X','XI','XIV']))]
print(ConoData_cyclic.shape)

ConoData_cyclic['Sequence'] = ConoData_cyclic['Sequence'].str.replace(" ","") #remove space in sequence
ConoData_cyclic.to_excel('ConoServer_cyclicPep_data.xlsx', index=False, header=True) #save to excel
```

```python
##***Extract cyclic peptides from ConoServer***##
structure_html = BeautifulSoup(open("ConoServer_structure.html"),'html.parser')
structure_tables = structure_html.find_all('table')
structure_table = structure_tables[1] ##id="datatable"
structure_data = pd.read_html(structure_table.prettify())
```

```python
structure_data[0].head()
```

```python
##remove row/column only contains NAs
print(structure_data[0].shape)
ConoData_structure = structure_data[0].dropna(axis=0,how='all')
ConoData_structure = ConoData_structure.dropna(axis=1,how='all')
print(ConoData_structure.shape)
```

```python
##save to excel
ConoData_structure.to_excel('ConoData_structure.xlsx', index=False, header=False)
```

```python
##data processing
ConoData_structure = pd.read_excel('ConoData_structure.xlsx', dtype={0:'str',18:'str'})
ConoData_structure['ID'] = 'S'+ConoData_structure['ID'].str.zfill(width=5) #fill structure ID
ConoData_structure['Biological Magnetic Resonance Data Bank'] = ConoData_structure['Biological Magnetic Resonance Data Bank'].replace('0', np.nan) #replace 0 with NaN
#ConoData_structure.to_excel('ConoData_structure_v2.xlsx', index=False, header=True) #save processed data
```

```python
##merge structure information and peptide information
ConoData_merge = pd.merge(ConoData_cyclic, 
                          ConoData_structure.loc[:,['ID','Name','Notes','Structure type','References','Full References','Biological Magnetic Resonance Data Bank','PDBcode','Last update']], 
                          how='left', left_on='Structure', right_on='Name',
                          suffixes=['_p','_s']
                         )
ConoData_merge.head()
```

```python
##Save to excel
ConoData_merge.to_excel('ConoServerData_merge_v1.xlsx', index=False, header=True)
```
