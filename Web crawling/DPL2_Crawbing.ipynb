```python
# DPL2_Crawbing
import requests  # 导入requests包
from urllib3.exceptions import InsecureRequestWarning
from urllib3 import disable_warnings
from bs4 import BeautifulSoup
from lxml import etree
import openpyxl

wb = openpyxl.Workbook()
sheet1 = wb.active
sheet1.append(
    ['DPL', 'name', 'Source', 'Squence', 'length', 'Struct_File', 'Struct_Source', 'physical charge', 'function',
     'Receptor', 'DOI', 'Infor', 'Receptor-structure', 'Kd', 'IC50', 'Receptor-function'])
sheet2 = wb.create_sheet('title')
sheet2.title = 'new sheet'
sheet1.sheet_properties.tabColor = '000000'
disable_warnings(InsecureRequestWarning)
if __name__ == '__main__':
    headers = {
        'User-Agent': "Mozilla/5.0 (Linux; Android 6.0; Nexus 5 Build/MRA58N) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/103.0.0.0 Mobile Safari/537.36"
    }
    url = 'http://www.peptide-ligand.cn/index.php/search-test/'
    response = requests.get(url=url, headers=headers, verify=False)
    response.encoding = response.apparent_encoding
    label = etree.HTML(response.text)
    tr_list = label.xpath('//*[@id="tablepress-9"]/tbody/tr')
    for tr in tr_list:
        charge, Struct_Source, Struct_Down, Struct_File, Infor, DOI, Receptor, Source, function = '', '', '', '', '', '', '', '', ''
        rece_struct, Kd, IC50, rece_function = '', '', '', ''
        STR_type = "".join(tr.xpath('./td[5]//text()'))
        if STR_type == "C" or STR_type == "Cyclic":
            DPL_id = "".join(tr.xpath('./td[1]/text()'))
            name = "".join(tr.xpath('./td[2]/text()'))
            Squence = "".join(tr.xpath('./td[3]/text()'))
            length = "".join(tr.xpath('./td[6]/text()'))
            charge = "".join(tr.xpath('./td[7]/text()'))
            Struct_Source = "".join(tr.xpath('./td[9]/text()'))
            Struct_Down = "".join(tr.xpath('./td[10]/a/@href'))
            Struct_File = "have" if Struct_Down else 'None'
            Infor = "".join(tr.xpath('./td[19]/text()'))
            DOI = "".join(tr.xpath('./td[22]/text()'))
            Receptor = "".join(tr.xpath('./td[11]/text()'))
            Source = "".join(tr.xpath('./td[4]/text()'))
            function = "".join(tr.xpath('./td[8]/text()'))
            rece_struct = "".join(tr.xpath('./td[20]/text()'))
            Kd = "".join(tr.xpath('./td[17]/text()'))
            IC50 = "".join(tr.xpath('./td[18]/text()'))
            rece_function = "".join(tr.xpath('./td[21]/text()'))
            xx_info = [DPL_id, name, Source, Squence, length, Struct_File, Struct_Source, charge, function, Receptor,
                       DOI, Infor, rece_struct, Kd, IC50, rece_function]
            print(STR_type, xx_info)
            sheet1.append(xx_info)
    wb.save('DPL.xlsx')
```
